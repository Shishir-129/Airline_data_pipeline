```python
# creating 'is_delayed' column to indicate if a flight is delayed or not
# 'arr_delay' column contains arrival delay in minutes
df['is_delayed'] = (df['arr_delay'] > 15).astype(int)
df['is_delayed']
```

## Generating Graphs

```python
df.columns
```

```python
# Graph 1 : Creating a heatmap to see the correlation between attributes

sns.set_theme(style='darkgrid')
plt.figure(figsize=(15,11))
corr = df.corr(numeric_only=True) #calculating correlation only for numeric columns
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title('Correlation Heatmap of delay factors')
plt.show()
```

```python
# Graph 2 : Histogram to visualize the distribution of arrival delays

plt.figure(figsize=(10,6))
sns.histplot(data=df,x=df[df['arr_delay']<200]['arr_delay'],kde=True,color='crimson')
plt.title("Distribution of Arrival Delays (in minutes)")
plt.xlabel("Arrival Delay (in minutes)",labelpad=15)
plt.show()
```

```python
# Graph 3 : Bar plot to compare average arrival delay for each carrier

plt.figure(figsize=(10,6))

delay_order = df.groupby('carrier')['arr_delay'].mean().sort_values(ascending=False)

# y-axis value is the mean(average) of arr_delay for each carrier i.e. just like df.groupby('carrier')['arr_delay'].mean()
sns.barplot(data=df,x='carrier',y='arr_delay',order=delay_order.index,hue='carrier',palette='viridis')
plt.ylabel("Arrival Delay (in minutes)")
plt.xlabel("Airline Carrier",labelpad=15)
plt.title("Average Arrival Delay by Airline Carrier",pad=15)
plt.show()
```

```python
# Graph 4 : Count plot to show the number of delayed flights for each carrier

plt.figure(figsize=(12,8))

delayed_df = df[df['is_delayed'] == 1]
sns.countplot(data=delayed_df,x='carrier',hue='carrier',order=delayed_df['carrier'].value_counts().index)
plt.xlabel("Airline Carrier",labelpad=15)
plt.ylabel("Number of Delayed Flights")
plt.title("Number of Delayed Flights by Airline Carrier",pad=15,fontsize=15)
plt.show()
```

```python
df.columns
```

```python
# Graph 5 : Box plot to visualize the distribution of delay causes

plt.figure(figsize=(10,5))

delay_causes = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']

sns.boxplot(data=df[delay_causes],palette='Set2')
plt.ylim(0,5000) #setting y-axis limit to focus on the main distribution and reduce the impact of extreme outliers
plt.title("Statistical Summary of Delay Causes")
plt.ylabel("Minutes of Delay")
plt.show()
```

In this plot we can see that most outliers are in the 'late_aircraft_delay' column, which indicates that delays caused by late aircraft are more variable and can be significantly higher than other delay causes. The 'weather_delay' column also shows some outliers, suggesting that weather-related delays can also be quite severe in certain cases. The other delay causes have fewer outliers, indicating that they tend to be more consistent in terms of delay duration.

In 'security_delay' column, the median is very close to zero, which suggests that most flights do not experience significant security-related delays. However, the presence of outliers indicates that there are occasional instances where security issues can lead to substantial delays.

```python
# Graph 6 : Scatter plot to explore the relationship between late aircraft delay and arrival delay

plt.figure(figsize=(12,8))

sns.scatterplot(data=df,x='late_aircraft_delay',y='arr_delay',hue='carrier')
plt.title("Relationship between Late Aircraft Delay and Arrival Delay")
plt.ylabel("Arrival Delay (in minutes)")
plt.show()
```

```python
# Graph 7 : Line plot to show the trend of average arrival delay over months

monthly_delays = df.groupby('month')['arr_delay'].mean().reset_index() 
monthly_delays
plt.figure(figsize=(12,6))

sns.lineplot(data=monthly_delays,x='month',y='arr_delay',marker='o',color='teal')
plt.xticks(range(1,13)) #setting x-axis ticks to represent each month
plt.ylabel("Arrival Delay (in mins)",labelpad=15)
plt.xlabel("Month : 1=Jan, 12=Dec",labelpad=15)
plt.title("Average Arrival Delay by Month",pad=15,fontsize=15)
plt.show()
```

## Summary of the dataset: ##

**1. Which airline is the King of Delays?**
   - Based on the BarPlot(graph 3) and CountPlot(graph 4) of delayed flights by airline carrier, the carrier with the highest number of delayed flights is the "WN" of 'SouthWest Airlines'. This indicates that Southwest Airlines has the most significant delay issues among the carriers in the dataset.

**2. Which delay factor is the most common?**
   - From the heatmap(graph 1) and the distribution of arrival delays(graph 2), it appears that 'late_aircraft_delay' is a significant factor contributing to flight delays. The scatter plot(graph 6) also shows a strong relationship between late aircraft delay and arrival delay, suggesting that late aircraft is a common and impactful delay factor in this dataset.

**3. Peak Delay Month?**
   - From the line plot (graph 7), we can see peak delays during June, July, and August. This suggests that the summer months experience higher delays, which could be due to increased travel demand, weather conditions, or other seasonal factors affecting flight operations during this period.

**4. Optimal Arrival Delay?**
   - From Barplot(graph 3) and Countplot(graph 4), we can see that 'OO' (SkyWest Airlines) has largest number of flight delays but the average arrival delay is very low compared to other carriers. This suggests that while SkyWest Airlines may have a higher frequency of delays, the severity of those delays in terms of arrival time is relatively low compared to other carriers.
